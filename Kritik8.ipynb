```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
h = 0.001
def multiply_list(list,k):
    ny=[]
    for i in range (0,len(list)):
        ny.append(list[i]*k)
    return ny
def derivative(f,c):
    return [((f(c[0]+h,c[1])-f(c[0]-h,c[1]))/(2*h)),((f(c[0],c[1]+h)-f(c[0],c[1]-h))/(2*h))] #Use the limit definition to define the derivative at any point
def new_point(f,c,a):
    placeholder = multiply_list(derivative(f,c),a) #define the new point as the gradient * the original point
    return [c[0]-derivative(f,c)[0]*a,c[1]-derivative(f,c)[1]*a]
def euclidean_distance(c1,c2): 
    sum = 0
    for i in range (0,len(c1)):
        sum+=(c1[i]-c2[i])**2
    return sum**0.5
def gradient(f,c,a, limit):
    for i in range (0,limit):
        c=new_point(f,c,a)
    return c
def f1(x,y):
    return x**2+y**2
def f2(x,y):
    return 1-np.exp(-x**2-(y-2)**2)-2*np.exp(-x**2-(y+2)**2)

print(gradient(f1,[0.1,0.1],0.1,10))
print(gradient(f1,[1,-1],0.01,100))

```

```python
import numpy as np
import matplotlib.pyplot as plt #usual matplotlib
from mpl_toolkits import mplot3d 

x=np.linspace(-5,5,100)
y=np.linspace(-5,5,100)
x,y=np.meshgrid(x,y)
z= f2(x,y)
fig = plt.figure()
ax = plt.axes(projection='3d')
ax.plot_surface(x, y, z,cmap='viridis', edgecolor='none')
plt.show()
#x,y z are variable names.
```
