```python
import matplotlib.pyplot as plt 
import numpy as np

#Define functions f(x) and g(x)
def f(x):
      return x**2 
def g(x):
    return 100*(x**3)+100*(x**2)-100*x+100 #funnier function in my opinion

def funny_function(x):
    if x>0:
        return x**x
    elif x==0:
        return 1
    else:
        return abs(x)**abs(x)
def deriv(f, base):
      return ((f(base+10**(-5))-f(base))/(10**(-5))) #First principles derivative format

def gradient_descent(f, L, x0):
    x_coords = []
    y_coords = []
    x_coords.append(x0) #setting first x coordinate to x0
    y_coords.append(f(x0)) #setting first y coordinate to f(x0)
    xi=x0
    i=1
    while abs(deriv(f,xi))>0.01:
        xi=x_coords[i-1]-L*deriv(f,x_coords[i-1]) #x1 = x0 + learning_value * f'(x)
        yi=f(xi)
        x_coords.append(xi) #add xi to x for last value
        y_coords.append(yi) #add yi to y for last value
        i=i+1



    plot_range = np.linspace(min(x_coords)-0.5,max(x_coords)+0.5,1000)
    f_range = [f(i) for i in plot_range]
    plt.plot(plot_range,f_range) #plot the range and function
    plt.plot(x_coords,y_coords) #plot the x and y coords of Euler's method
    plt.title("Gradient Descent Visualization")
    plt.xlabel("x")
    plt.ylabel("f(x)")
    plt.show()
    print("the final x coordinate is x =", x_coords[i-1])

gradient_descent(f,0.9,5)
```
