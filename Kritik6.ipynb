```python
import numpy as np
import matplotlib.pyplot as plt
def fx(mu, sig, x):
    return ((2*np.pi*(sig))**-0.5)*np.exp(-((x-mu)**2)/(2*(sig))) # density function

def integration(fx,a,b):
    n=1000 # number of iterations
    x=np.linspace(a,b,n) # make n many values of x, evenly spaced so they can be used for the integration 
    y_middle = []
    delta_x = (b-a)/n 
    for i in range(n-1):
        xm = (x[i]+x[i+1])/2 # middle of the last and next point
        ym = fx(171,7.1**2, xm) # standard deviation and mean for the average male height, see it as f(x) at xi
        y_middle.append(ym) # sum of the y values x the interval in x (ie y1x + y2x = x(y1+y2))
    ms = sum(y_middle)*delta_x
    return ms 
print(integration(fx,162,190))
0.8929122447004852
x = np.linspace(-10,10,100)
fx_values = fx(0,5,x)
plt.plot(x,fx_values, label = 'normal distribution')
plt.show()
```
